## Observations and Insights 

```python
'''
From this data it appears that Capomulin and Ramicane seem to reduce tumor volume while the mice 
using the drugs Infubinol and Ceftamin appeared to have tumor size increase overall. There seems to be 
a significant positive coorelation between tumor volume and mouse weight. 


'''
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import sklearn.datasets as dta
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata, study_results, on='Mouse ID')

# Display the data table for preview
combined_data.head(50)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)


capomulin_tumor_vol = final_tumor_volume.loc[final_tumor_volume["Drug Regimen"] == 'Capomulin', :]
ramicane_tumor_vol = final_tumor_volume.loc[final_tumor_volume["Drug Regimen"] == 'Ramicane', :]
infubinol_tumor_vol = final_tumor_volume.loc[final_tumor_volume["Drug Regimen"] == 'Infubinol', :]
ceftamin_tumor_vol = final_tumor_volume.loc[final_tumor_volume["Drug Regimen"] == 'Ceftamin', :]


tumor_vol = [capomulin_tumor_vol, ramicane_tumor_vol, infubinol_tumor_vol, ceftamin_tumor_vol]



capomulin_iqr = capomulin_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.75)-capomulin_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.25)
ramicane_iqr = ramicane_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.75)-ramicane_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.25)
infubinol_iqr = infubinol_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.75)-infubinol_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.25)
ceftamin_iqr = ceftamin_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.75)-ceftamin_tumor_vol['Tumor Volume (mm3)'].quantile(q=0.25)

iqr = [capomulin_iqr, ramicane_iqr, infubinol_iqr, ceftamin_iqr]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
capomulin_outliers = capomulin_tumor_vol.loc[(capomulin_tumor_vol["Tumor Volume (mm3)"] < (capomulin_tumor_vol['Tumor Volume (mm3)'].mean() - (1.5 * capomulin_iqr)))
                        | (capomulin_tumor_vol["Tumor Volume (mm3)"] > capomulin_tumor_vol['Tumor Volume (mm3)'].mean() + 1.5 * capomulin_iqr),
                        :]

ramicane_outliers = ramicane_tumor_vol.loc[(ramicane_tumor_vol["Tumor Volume (mm3)"] < (ramicane_tumor_vol['Tumor Volume (mm3)'].mean() - (1.5 * ramicane_iqr)))
                        | (ramicane_tumor_vol["Tumor Volume (mm3)"] > ramicane_tumor_vol['Tumor Volume (mm3)'].mean() + 1.5 * ramicane_iqr),
                        :]

infubinol_outliers = infubinol_tumor_vol.loc[(infubinol_tumor_vol["Tumor Volume (mm3)"] < (infubinol_tumor_vol['Tumor Volume (mm3)'].mean() - (1.5 * infubinol_iqr)))
                        | (infubinol_tumor_vol["Tumor Volume (mm3)"] > infubinol_tumor_vol['Tumor Volume (mm3)'].mean() + 1.5 * infubinol_iqr),
                        :]

ceftamin_outliers = ceftamin_tumor_vol.loc[(ceftamin_tumor_vol["Tumor Volume (mm3)"] < (ceftamin_tumor_vol['Tumor Volume (mm3)'].mean() - (1.5 * ceftamin_iqr)))
                        | (ceftamin_tumor_vol["Tumor Volume (mm3)"] > ceftamin_tumor_vol['Tumor Volume (mm3)'].mean() + 1.5 * ceftamin_iqr),
                        :]

print('Capomulin Outliers')
print(capomulin_outliers)
print('-----------------')
print('Ramicane Outliers')
print(ramicane_outliers)
print('-----------------')
print('Infubinol Outliers')
print(infubinol_outliers)
print('-----------------')
print('Ceftamin Outliers')
print(ceftamin_outliers)


#No outliers for Ceftamin



    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_tumor = combined_data.loc[combined_data["Drug Regimen"] == 'Capomulin', :]
capomulin_tumor_groupby = capomulin_tumor.groupby('Mouse ID')
capomulin_average = capomulin_tumor_groupby.mean()


plt.scatter(capomulin_average['Weight (g)'], capomulin_average['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume vs Mouse Weight')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 

correlation = round(st.pearsonr(capomulin_average['Weight (g)'], capomulin_average['Tumor Volume (mm3)'])[0], 2)
print(f'Correlation coefficient between average tumor volume and mouse weight: {correlation}')

(slope, intercept, rvalue, pvalue, stderr) = linregress(capomulin_average['Weight (g)'], capomulin_average['Tumor Volume (mm3)'])
regress_values = capomulin_average['Weight (g)'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print("Linear regression model: y = " + str(round(slope,2)) + "x + " + str(round(intercept,2)))


plt.scatter(capomulin_average['Weight (g)'], capomulin_average['Tumor Volume (mm3)'])
plt.plot(capomulin_average['Weight (g)'],regress_values,"r-")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume vs Mouse Weight')
plt.annotate(line_eq,(20,36),fontsize=15,color="red")

plt.show()
```
